```python
import pandas as pd

path = 'cars_cleaned.csv'
df_ = pd.read_csv(path)
df = df_.copy()
```

```python
df= df[df['Price']< 20000]
```

```python
df_ref = pd.read_excel('vehicle_type.xlsx')
df_ref
```

```python
df_ref['vehicle_type'].value_counts()
```

```python
df['vehicle_type'] = ''

for model, v_type in zip(df_ref['brand_model_names'], df_ref['vehicle_type']):
    df.loc[df['name'].str.contains(model, na=False), 'vehicle_type'] = v_type

df = df.reset_index(drop=True)
df
```
