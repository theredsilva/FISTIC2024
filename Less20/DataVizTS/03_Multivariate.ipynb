# Multivariate Time Series Manipulation & Visualization

```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

path = 'energy_generation_hourly_year.csv'

df = pd.read_csv(path,
                 index_col=0, 
                 parse_dates=True,
                 )
df
```

```python
df.plot(figsize=(15,8));
```

## Resampling the Time
- [Date offset `rule` list](https://pandas.pydata.org/docs/user_guide/timeseries.html#dateoffset-objects)
- [Pandas plot `kind` list](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)

### Daily

```python
df.resample(rule='D').sum().plot(figsize=(15,8))
```

### Monthly

```python
df.resample(rule='M').sum().plot(figsize=(15,8))
```

```python
df.resample(rule='M').sum().plot.area(figsize=(15,8))
```

```python
df.resample(rule='M').sum().plot.bar(figsize=(15,8));
```

## Visualize Time Series Correlation

```python
df_weekly = df.resample('7D').sum()
df_weekly
```

```python
df_weekly.plot(kind='scatter', x='Solar Photovoltaic', y='Eolic');
```

```python
df_weekly.plot(kind='scatter', x='Solar Photovoltaic', y='Hydraulic');
```

## Pivot Tables

### Transform object to DataFrame

> Only if you don't have a DataFrame yet.

```python
df
```

### Time series categories in one column

```python
dff = df.melt(ignore_index=False, value_name='generation', var_name='technology')
dff
```

### Create columns with datetime information

```python
dff = (dff
        .assign(
            year=lambda df: df.index.year,
            month=lambda df: df.index.month,
            day=lambda df: df.index.day,
            hour=lambda df: df.index.hour))

dff
```

### Process `pivot_table`

- Use this methodology.

```python
dff = dff.pivot_table(
                        values='generation',
                        index='technology',
                        columns=['year', 'month'],
                        aggfunc='sum')

dff
```

### Style DataFrame

```python
(dff
 .div(1_000_000)
 .sort_values(by=(2023,1), ascending=False)
 .style
    .background_gradient(cmap='Oranges', axis=None)
    .format(precision=2))
```

```python
dff = (dff
        .div(1_000_000)
        .sort_values(by=(2023,1), ascending=False)
        .style
        .background_gradient(cmap='Oranges', axis=0)
        .format(precision=2))

dff
```
